prepare the data by doing: 
*checking all columns that have missing values
*One-Hot Encoding for Categorical variables such as Club, Nationality, Preferred Positions

```python
# Data Preparation Step 1: check how many missing values are in the dataset
BlackFriday_Dataset.isnull().sum()
```

it looks like there are missing value in category 1&2 which means no purchasing in this items. Lets zero it then. 

```python
# Data Preparation Step 2: Fill the missing cell with zero
BlackFriday_Dataset.fillna(0)
```

```python
# Data Preparation Step 3: One-Hot Encoding for Categorical variables
# One-hot encode the feature:  Gender, Age, City_Category, Stay_in_curent_City_Years
le = LabelEncoder()
BlackFriday_Dataset['Gender_onehot_encode']              = le.fit_transform(BlackFriday_Dataset['Gender'].astype(str))
BlackFriday_Dataset['Age_onehot_encode']                 = le.fit_transform(BlackFriday_Dataset['Age'].astype(str))
BlackFriday_Dataset['City_Category_onehot_encode']       = le.fit_transform(BlackFriday_Dataset['City_Category'].astype(str))
BlackFriday_Dataset['Stay_In_Current_City_Years_encode'] = le.fit_transform(BlackFriday_Dataset['Stay_In_Current_City_Years'].astype(str))
```

lets start to answers our questions. 

```python
# Question 1:Customers more likely to spend during black firday?top 20.
plt.figure(figsize = (20,8))
BlackFriday_Dataset.groupby('User_ID')['Purchase'].sum().nlargest(20).sort_values().plot('barh')
```

this will help to understand most attracted costumers that willing to purchase, that will help decision maker to act accordingly. 

```python
# Question 2: costumers ages are more likly attracted in that day and which gender? 
plt.figure(figsize = (20,8))
sns.countplot(BlackFriday_Dataset['Age'])
```

```python
plt.figure(figsize = (20,8))
sns.countplot(BlackFriday_Dataset['Age'],hue=BlackFriday_Dataset['Gender'])
```

its very clear that , the age between 26-35 that have higher income with young spirit more likely to shop.Then age of 18-25 comes second. 

```python
# Question 3: items has more sales? top 20.
plt.figure(figsize = (20,8))
BlackFriday_Dataset.groupby('Product_ID')['Purchase'].count().nlargest(20).sort_values().plot('barh')
```

By the above plot, the sellers have better pictures to see most demanded items.

```python
# Question 4: cities are more active than others? 
plt.figure(figsize = (20,8))
sns.countplot(BlackFriday_Dataset['Occupation'], hue = BlackFriday_Dataset["City_Category"])
```

In the above plot, its clear that the shopping in city B is more than A&C due to the size of the city B larger and has more population than others. Although the occupation of 0&4&7 got more shoppers.
